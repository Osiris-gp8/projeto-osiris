```python
import requests
import json
import pandas as pd
```

```python
cidades_json = json.loads(requests\
                          .get("https://servicodados.ibge.gov.br/api/v1/localidades/municipios").text)
cidades_json
```

```python
cidades_list = []
for cidade in cidades_json:
    cidades_list.append(
    {
        'id':cidade['id'],
        'UF': cidade['microrregiao']['mesorregiao']['UF']['sigla'],
        'nome': cidade['nome']
    }
    )
df_cidades = pd.DataFrame(data=cidades_list)
cidades_list
```

```python
df_cidades.sample(n=10000, replace=True)
```
